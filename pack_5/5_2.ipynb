```python
import random
random.seed()
t = 0
x_data = []
y_data = []
n_data = []
print("Введите число симуляций: ")
n = int(input())
for j in range(n):
    f = False
    print("День №", j + 1)
    print("Клавдия тестирует код...")
    for i in range(5):
        k = random.randint(1, 100)
        if k <= 11:
            f = True
            print("Прошло ", (i + 1) * 17, " минут.Клавдия выявила баг.")
            print("Встречи с Константином сегодня точно не будет, так как баг будет исправлен.")
            break
        else: 
            print("Прошло ", (i + 1) * 17, " минут. Клавдия не выявила баг")
    if not f:   
        k = random.randint(1, 100)
        if k <= 11:
            f = True
            t += 1
            print("Прошло ", 6 * 17, " минут. Будет встреча с Константином ;)")
            continue
    else: 
        y_data.append(t / (j+1))
        x_data.append(j)
        n_data.append(((0.89)**5)*0.11)
        continue
    y_data.append(t / (j+1))
    x_data.append(j)
    n_data.append(((0.89)**5)*0.11)
    print("Прошло ", 6 * 17, " минут. Константин не заметил баг. Не будет встречи с Константином :с")
print("Вероятность встречи с Константином: ", t / n)
```

Вычислим вероятность встречи Иннокентия с Константином. Для этого Клавдия должна ни разу за полтора часа тестирования не встретить баг, то есть вероятность этого $(1-0.11)^5$. Далее баг должен встретить Константин и вероятность совокупности этих событий  $0.89^5*0.11$.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.title("Вероятность от количества итераций")
plt.plot(x_data, y_data, color = 'red') #результат симуляции
plt.plot(x_data, n_data, color  = 'green') #теоретически вычисленное значение 
```

Как видно из графиков, с ростом числа симуляций график, использующий эсперементально полученные значения все больше приближается к теоретически полученному значению $0.0614$.
