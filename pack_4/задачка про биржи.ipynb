Пусть d - словарь, где ключом будет название биржи, а значением - структура данных типа "очередь".
Произведем линейный проход по файлу. Раз за разом считывая строку, будем извлекать из нее название биржи и время сделки.
Таким образом за каждый проход мы будем взаимодействовать только с одним объектом из d и только одной строчкой из файла.
Для того, чтобы определить входит ли новая сделка в допустимый промежуток времени, сделаем также линейный цикл, где каждый первый элемент очереди будет проверяться на вхождение в данный промежуток. При этом мы будем проверять ранее добавленные в очередь элементы, постепенно убирая их из нее, если они не подходят промежутку. То есть по каждому элементу мы пройдемся в среднем только пару раз.  Все остальные действия сравнимы с $O(1)$, из-за чего можно сделать вывод, что время работы программы сравнимо с  $O(2*n) = O(n)$.
